```python
# Viết một hàm tinh_tong(n) nhận vào một số nguyên dương n sau đó tính tổng các số từ 1 đến n.
def tinhtong(n):
    tong = 0
    for i in range(1,n+1):
        tong = tong + i
    return tong

tinhtong(8)
```

```python
# Viết một hàm tinh_tong_chan(n) nhận vào một số nguyên dương n sau đó tính tổng các số chẵn từ 1 đến n.
def tinhtongchan(n):
  tongchan = 0
  for i in range(2, n+1, 2):
      tongchan = tongchan+i
  return tongchan

tinhtongchan(5)
```

```python
# Viết một hàm tim_so_lon_nhat(danh_sach) nhận vào một danh sách chứa 10
# số được nhập vào từ bàn phím. Sau đó trả về số lớn nhất trong danh sách

danh_sach = []
for i in range(5):
    so = int(input("Nhap so thu " + str(i+1) + ": "))
    danh_sach.append(so)

def tim_so_lon_nhat(danh_sach):
    so_lon_nhat = danh_sach[0]
    for so in danh_sach:
        if so > so_lon_nhat:
            so_lon_nhat = so
    return so_lon_nhat

print("Dãy số được nhập: {0}".format(danh_sach))
print("Số lớn nhất trong dãy số là : {0}".format(tim_so_lon_nhat(danh_sach)))

print("Giá trị max:",max(danh_sach)) # Hàm max() trả về kết quả lớn nhất trong 1 danh sách
print("Giá trị min:",min(danh_sach))
```

```python
# Viết một hàm tim_so_be_nhat(danh_sach) nhận vào một danh sách chứa 10
# số được nhập vào từ bàn phím. Sau đó trả về số bé nhất trong danh sách.

def tim_so_be_nhat(danh_sach):
    so_be_nhat = danh_sach[0]
    for so in danh_sach:
        if so < so_be_nhat:
            so_be_nhat = so
    return so_be_nhat
print("Số bé nhất trong dãy số là : {0}".format(tim_so_be_nhat(danh_sach)))
```

```python
# Viết một hàm kiem_tra_so_nguyen_to(n) nhận vào một số nguyên dương n và kiểm tra xem n có phải là số nguyên tố hay không
def kiem_tra_so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True


n = int(input("Nhập n: "))
if kiem_tra_so_nguyen_to(n):
    print(n, "là số nguyên tố")
else:
    print(n, "không phải là số nguyên tố")
```

```python
# Viết một hàm dem_so_chu_so(n) nhận vào một số nguyên dương n và đếm số chữ số của n

def dem_so_chu_so(n):
    count = 0
    while n > 0:
        n = n // 10
        count += 1
    return count

n = int(input("Nhập n: "))
so_chu_so = dem_so_chu_so(n)
print("Số chữ số của", n, "là", so_chu_so)
```

```python
# Nhập vào lần lượt 3 hệ số a, b, c của phương trình bậc 2 (ax^2+bx+c = 0). Giải phương trình bậc 2 trên và in nghiệm ra màn hình.
import math

a = float(input("Nhập hệ số a: "))
b = float(input("Nhập hệ số b: "))
c = float(input("Nhập hệ số c: "))

delta = b**2 - 4*a*c

if delta < 0:
    print("Phương trình vô nghiệm")

elif delta == 0:
    x = -b / (2*a)
    print("Phương trình có nghiệm kép x =", x)

else:
    x1 = (-b + math.sqrt(delta)) / (2*a)
    x2 = (-b - math.sqrt(delta)) / (2*a)
    print("Phương trình có hai nghiệm phân biệt:")
    print("x1 =", x1)
    print("x2 =", x2)
```

```python
# Tạo 3 biến lưu trữ 3 giá trị nhập vào ngày tháng năm, in ra ngày hôm sau và ngày hôm qua của ngày vừa nhập. (Bài tập khó)

import datetime

day = int(input("Nhập ngày: "))
month = int(input("Nhập tháng: "))
year = int(input("Nhập năm: "))

# Tạo đối tượng datetime từ ngày tháng năm
date_input = datetime.datetime(year, month, day)

# In ra ngày hôm sau và ngày hôm qua
date_tomorrow = date_input + datetime.timedelta(days=1)
date_yesterday = date_input - datetime.timedelta(days=1)

print("Ngày hôm sau là:", date_tomorrow.strftime("%d/%m/%Y"))
print("Ngày hôm qua là:", date_yesterday.strftime("%d/%m/%Y"))
```

```python
# Viết một hàm kiem_tra_so_nguyen_to(n) nhận vào một số nguyên dương n và kiểm tra xem n có phải là số nguyên tố hay không.
#            Sử dụng vòng lặp và kiểm tra các ước số của n để thực hiện tác vụ này.

def kiem_tra_so_nguyen_to(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

n = int(input("Nhập n: "))
if kiem_tra_so_nguyen_to(n):
    print(n, "là số nguyên tố")
else:
    print(n, "không phải là số nguyên tố")
```

```python
# Viết một hàm tinh_giai_thua(n) nhận vào một số nguyên dương n và tính giai thừa của n (n!)

def tinh_giai_thua(n):
    giai_thua = 1
    if n < 0:
        return "Không tồn tại giai thừa cho số âm"
    elif n == 0:
        return "Giai thừa của 0 là 1"
    else:
        for i in range(1, n + 1):
            giai_thua = giai_thua * i
        return giai_thua

n = int(input("Nhập n: "))
tinh_giai_thua(n)
```

```python
# Viết một hàm tinh_tong(n) nhận vào một số nguyên dương n và tính tổng các chữ số của n. Ví dụ: n = 123, thì hiển thị kết quả là 6.
def tinh_tong(n):
    tong = 0
    while(n > 0):
        tong = tong + (n % 10)
        n = n // 10
    return tong

n = int(input("Nhập n: "))
tinh_tong(n)
```

```python
# Viết một hàm tinh_trung_binh(data) nhận vào một danh sách số và tính trung bình cộng của các số trong danh sách đó.
#            Sử dụng vòng lặp để lặp qua từng số trong danh sách và tính tổng các số
def tinh_trung_binh(data):
    tong = 0
    for so in data:
        tong += so
    if len(data) > 0:
        trung_binh = tong / len(data)
        return trung_binh
    else:
        return None

data = [1,2,3,4,2,3,4,5,6]
tinh_trung_binh(data)
```

```python
# Viết một hàm dem_so_lan_xuat_hien(data) nhận vào một danh sách và đếm số lần xuất hiện của mỗi phần tử trong danh sách đó.
def dem_so_lan_xuat_hien(data):
    dem = {}
    for i in data:
        if i in dem:
            dem[i] += 1
        else:
            dem[i] = 1
    return dem

dem_so_lan_xuat_hien([1,2,1,1,2,3,'x','x','x','y'])
```
