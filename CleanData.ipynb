```python
# Load file bang thu vien pandas, luu o dataframe
infobrasil = pd.read_csv("/content/brasil-real-estate-1.csv")
```

```python
#Xem thông tin tổng quan về bộ dữ liệu
infobrasil.info()
# object = string
```

```python
#Xem tổng quan về bộ dữ liệu
infobrasil.head(10)
```

```python
#Xoá bỏ các dữ liệu bị trống
infobrasil.dropna(inplace=True)
```

```python
infobrasil.reset_index()
```

```python
infobrasil.info()
```

```python
#Cleaning dữ liệu - Tách cột lattitude - longtitude thành 2 cột riêng
infobrasil[["lat", "lon"]] = infobrasil["lat-lon"].str.split(",", expand = True)
#expand tạo 2 cột, split chia tách 2 cột lat và lon
```

```python
infobrasil
```

```python
infobrasil['lat'] = infobrasil['lat'].astype(float)
infobrasil['lon'] = infobrasil['lon'].astype(float)
```

```python
#đổi kiểu dữ liệu về dạng số, sau đó kiểm tra lại.
infobrasil[["lat", "lon"]]= infobrasil[["lat", "lon"]].astype(float)
infobrasil[["lat", "lon"]].info()
```

```python
infobrasil.drop(columns = ["lat-lon"], inplace = True)
# xóa những cột không cần dùng
```

```python
infobrasil.head(10)
```

```python
# Cột price_usd dữ liệu vẫn đang lộn xộn, có cả kí tự "," để biểu thị hàng nghìn, do vậy ta sẽ xoá kí tự , đi.
# Kết hợp đó, xoá cả kí tự $ và đổi về dạng số thập phân

infobrasil["price_usd"] = infobrasil["price_usd"].str.replace("$","").str.replace(",","").astype(float)
infobrasil.info()
```

```python
infobrasil.head(5)
```

```python
# Cột "place_with_parent_names" cũng chưa đạt chuẩn dữ liệu.
# Tạo 1 cột mới đặt tên là state và sau đó tách cột place_with_parent_names này ra và lấy thông tin tên tỉnh của dữ liệu đó.
# Tên tỉnh nằm ở vị trí thứ 2 trong chuỗi.

infobrasil["state"] = infobrasil["place_with_parent_names"].str.split("|", expand = True)[2]
infobrasil.head(5)
```

```python
infobrasil.drop(columns = ["place_with_parent_names"], inplace = True)
infobrasil.head(5)
```

```python
file = infobrasil.to_csv("file_cleaned.csv")
```
